```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def getFrequency(note):
    A4=440
    notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

    octave = int(note[2]) if len(note) == 3 else int(note[1])
        
    keyNumber = notes.index(note[0:-1]);
    
    if (keyNumber < 3) :
        keyNumber = keyNumber + 12 + ((octave - 1) * 12) + 1; 
    else:
        keyNumber = keyNumber + ((octave - 1) * 12) + 1; 

    return A4 * 2** ((keyNumber- 49) / 12)
```

```python
amp = 1; sr = 10000; phase = 0

A5 = getFrequency('A5') # 라
B5 = getFrequency('B5') # 시
A4  = getFrequency('A4')
C5  = getFrequency('C5') # 도 
C6 = getFrequency('C6') # 높은 도
D5  = getFrequency('D5') # 레
E5  = getFrequency('E5') # 미 
Eb5 = getFrequency('D#5') 
F5  = getFrequency('F5') # 파
G5  = getFrequency('G5') # 솔
G4 = getFrequency('G4') # 낮은 솔
A4 = getFrequency('A4') # 낮은 라
B4 = getFrequency('B4') # 낮은 시
E4 = getFrequency('E4') # 낮은 미
F4 = getFrequency('F4') # 낮은 파
Bb5 = getFrequency('A#5')
D6 = getFrequency('D6')
```

```python
#notes = [C5, D5, E5, F5, G5, A5, B5, C6]

notes = [G5, E5, F5, G5, E5, F5, G5, G4, A4, B4, C5, D5, E5, F5,
         E5, C5, D5, E5, E4, F4, 
         G4, A4, G4, F4, G4, C5, B4, C5,
         C5, G5, C6]

durs = [0.7, 0.35, 0.35, 0.7, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35,
        0.7, 0.35, 0.35, 0.7, 0.35, 0.35, 
        0.35, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35, 0.35,
        0.15, 0.15, 0.7]
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
