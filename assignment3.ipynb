```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
df1 = pd.read_csv('/content/gdrive/MyDrive/assignment_yourgoodpants/yourgoodpants.csv', sep=',')
print(df1)
```

```python
df1.loc[df1.Time_s == "--undefined--", "Time_s"] = None
df1.loc[df1.F0_Hz == "--undefined--", "F0_Hz"] = None
print(df1)
```

```python
time = df1.Time_s
pitch = df1.F0_Hz

plt.plot(time, pitch, 'bo', label='pitch(Hz)')
plt.xlabel('Time(s)')
plt.ylabel('pitch(Hz)')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
```

```python
df2 = pd.read_csv('/content/gdrive/MyDrive/assignment_yourgoodpants/yourgoodpants_intensity.csv', sep=',')
print(df2)
```

```python
time = df2.Time_s
intensity = df2.Intensity_dB

plt.plot(time, intensity, 'ro', label = 'intensity(dB)')
plt.xlabel('Time(s)')
plt.ylabel('Intensity(dB)')
plt.legend(loc='best', ncol=2, fontsize=13, frameon=True, shadow=True)
plt.show()
```

```python
df3 = pd.read_csv('/content/gdrive/MyDrive/assignment_yourgoodpants/yourgoodpants_formants.csv', sep=',')
print(df3)
```

```python
time = df3.Time_s
F1 = df3.F1_Hz
F2 = df3.F2_Hz
F3 = df3.F3_Hz
F4 = df3.F4_Hz

plt.rcParams["figure.figsize"] = (10, 5)
plt.plot(time, F1, 'ro', label='F1')
plt.plot(time, F2, 'bo', label='F2')
plt.plot(time, F3, 'go', label='F3')
plt.plot(time, F4, 'yo', label='F4')
plt.xlabel('Time(s)')
plt.ylabel('Frequency(Hz)')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
```
